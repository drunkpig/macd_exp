```python
from futu import *
import pandas as pd
import numpy as np
from macdlib import *
```

```python
quote_ctx = OpenQuoteContext(host='futuapi.mkmerich.com', port=54012)
ret, data, page_req_key = quote_ctx.request_history_kline('SH.600478', start='2019-04-01', end='2019-08-02',
                                                          ktype=KLType.K_15M, fields=[KL_FIELD.DATE_TIME, KL_FIELD.CLOSE], max_count=2000) 
```

```python

data.to_csv('SH.600478.K_15M.csv')

df = pd.read_csv('SH.600478.K_15M.csv', index_col='time_key')
```

```python


df = pd.read_csv("SH.600478.K_15M.csv",index_col=0)
diff, dem, bar = macd(df)
df['macd_bar'] = bar
df['ma5'] = ma(df, 5)
df['ma10'] = ma(df, 10)
df['em_bar'] = (df['ma5']-df['ma10']).apply(lambda val: round(val, 2))


df30 = pd.read_csv("SH.600478.K_30M.csv",index_col=0)
diff, dem, bar = macd(df30)
df30['macd_bar'] = bar
df30['ma5'] = ma(df30, 5)
df30['ma10'] = ma(df30, 10)
df30['em_bar'] = (df30['ma5']-df30['ma10']).apply(lambda val: round(val, 2))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


def draw(df, start, end):
    fig, ax = plt.subplots(1,1, figsize=(20, 7))

    em_ax = ax
    em_ax.stem(df[start:end].time_key, df[start:end].em_bar, use_line_collection=True)
    em_ax.xaxis.set_tick_params(rotation=45)
    
    fig, ax = plt.subplots(1,1, figsize=(20, 7))

    em_ax = ax
    em_ax.stem(df[start:end].time_key, df[start:end].macd_bar, use_line_collection=True)
    em_ax.xaxis.set_tick_params(rotation=45)


draw(df, 1300, 1359)
draw(df30, 662, 679)


    
```
