```python

data.to_csv('SH.600478.K_15M.csv')

df = pd.read_csv('SH.600478.K_15M.csv', index_col=0)
```

```python
df['test'] = 0

seri = df.iloc[0:3].close

for i, close in seri.items():
    print(f'{i} -> {close}')

# 扫描除去零碎的绿柱区间
# 找到主峰，然后左右找到延伸。记录主峰
# 除去主峰（属于主峰的赋值为0）
# 当主峰高度为0的时候退出

min_val = df.iloc[15:40].close.min()
min_index = df.iloc[15:40].close.idxmin( axis=0)

df.at[min_index, 'test'] = 1
arr = df.iloc[15:40].close.array
type(arr)
# 从min_index先向左侧扫
arr_min_index = min_index - 15
for i in range(arr_min_index, 0+2, -1):
    
    if arr[i]<=arr[i-1] or arr[i]<=arr[i-2]:
        continue
    else:
        break # i 就是左侧波谷
        

# 从min_index向右侧扫描
for j in range(arr_min_index, 40-15-2):
    
    if arr[j]<=arr[j+1] or arr[j]<=arr[j+2]:
        continue
    else:
        break # j 就是右侧波谷
        
# 从[i, j]区间剩下的段继续加入迭代区间
print(f'===> {i} -> {j}')
```
